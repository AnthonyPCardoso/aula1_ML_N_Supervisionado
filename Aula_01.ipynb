```python
from sklearn.datasets import load_iris
# as_frame Se for True, os dados são um DataFrame do pandas incluindo colunas com dtypes apropriados (numéricos).
X, y = load_iris(return_X_y=True , as_frame=True)
```

```python
X = X.loc[:, ['petal length (cm)', 'petal width (cm)']]
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.scatter(X['petal length (cm)'], X['petal width (cm)'])
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2, random_state=0).fit(X)
```

```python
kmeans.labels_
```

```python
fig, ax = plt.subplots()
ax.scatter(X['petal length (cm)'], X['petal width (cm)'], c=kmeans.labels_)

plt.show()
```

```python
valores_k = []
inercias = []

for i in range(1,15):
  kmeans2 = KMeans(n_clusters=i,random_state=0).fit(X)
  valores_k.append(i)
  inercias.append(kmeans2.inertia_)
```

```python
fig, ax = plt.subplots()

ax.plot(valores_k, inercias)

plt.show()
```

```python
fig, ax = plt.subplots(ncols=2, figsize=(15, 6))

ax[0].scatter(X['petal length (cm)'],  X['petal width (cm)'], c=kmeans.labels_)
ax[1].scatter(X['petal length (cm)'],  X['petal width (cm)'], c=y)
```

```python
from sklearn import metrics

metrics.adjusted_rand_score(y, kmeans.labels_)
```

```python
kmeans3 = KMeans(n_clusters=3, random_state=0).fit(X)
```

```python
metrics.adjusted_rand_score(y, kmeans3.labels_)
```

```python
kmeans4 = KMeans(n_clusters=4, random_state =0).fit(X)
```

```python
metrics.adjusted_rand_score(y, kmeans4.labels_)
```

```python
valores_k =[]
ARI = []
RI = []

for i in range(1,15):
  kmeans = KMeans(n_clusters=i, random_state=0).fit(X)
  valores_k.append(i)
  RI.append(metrics.rand_score(y, kmeans.labels_))
  ARI.append(metrics.adjusted_rand_score(y,kmeans.labels_))
```

```python
fig, ax = plt.subplots()

ax.plot(valores_k,ARI)
ax.plot(valores_k,RI)

plt.show()
```

```python
valores_k = []
s = []

for i in range(2,15): # Começa com o número de clusters
  kmeans = KMeans(n_clusters=i,random_state=0).fit(X)
  valores_k.append(i)
  s.append(metrics.silhouette_score(X,kmeans.labels_))
```

```python
fig, ax = plt.subplots()

ax.plot(valores_k, s)

plt.show()
```
